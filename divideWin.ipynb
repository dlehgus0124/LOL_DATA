```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv("trans_position_allTiers.csv")

# 'win'를 포함하는 헤더만 필터링
filtered_headers = [col for col in df.columns if "win" in col]

# 필터링된 헤더의 열만 포함하는 DataFrame 생성
filtered_win_df = df[filtered_headers]

filtered_win_df.rename(columns={
    'win_champ': 'champ_code', 
    'win_tags': 'position_tags',
    'win_statperks': 'statperks',
    'win_styles_primary': 'styles_primary',
    'win_styles_sub': 'styles_sub',
    'win_items': 'items',
    'win_dealt': 'dealt',
    'win_taken': 'taken',
    'win_kills': 'kills',
    'win_deaths': 'deaths',
    'win_assists': 'assists',   
    'win_heals': 'heals',
    'win_cs': 'cs',
    'win_vision': 'vision',  
    }, inplace=True)

filtered_win_df['win'] = 1

filtered_win_df
```

```python
filtered_headers = [col for col in df.columns if "lose" in col]

# 필터링된 헤더의 열만 포함하는 DataFrame 생성
filtered_lose_df = df[filtered_headers]

filtered_lose_df.rename(columns={
    'lose_champ': 'champ_code', 
    'lose_tags': 'position_tags',
    'lose_statperks': 'statperks',
    'lose_styles_primary': 'styles_primary',
    'lose_styles_sub': 'styles_sub',
    'lose_items': 'items',
    'lose_dealt': 'dealt',
    'lose_taken': 'taken',
    'lose_kills': 'kills',
    'lose_deaths': 'deaths',
    'lose_assists': 'assists',   
    'lose_heals': 'heals',
    'lose_cs': 'cs',
    'lose_vision': 'vision',  
    }, inplace=True)

filtered_lose_df['win'] = 0

filtered_lose_df
```

```python
import pandas as pd

# 두 DataFrame 병합
merged_df = pd.concat([filtered_win_df, filtered_lose_df], ignore_index=True)

# 병합된 DataFrame을 CSV 파일로 저장
merged_df.to_csv("win_divide_data.csv", index=False)
```
